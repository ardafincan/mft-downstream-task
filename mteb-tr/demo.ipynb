```python
from sentence_transformers import SentenceTransformer
from huggingface_hub import login

login(token="hf_cBjj")

s_model = SentenceTransformer("magibu/embeddingmagibu-152m")
s_sentences = [
    "Bu mutlu bir adamdır",
    "Bu mutlu bir kadın",
    "Akşam sinemaya gideceğim",
    "Bugün hava güneşli"
]
s_embeddings = s_model.encode(s_sentences)

s_similarities = s_model.similarity(s_embeddings, s_embeddings)

f_model = SentenceTransformer("final_model")
f_embeddings = f_model.encode(s_sentences)
f_similarities = f_model.similarity(f_embeddings, f_embeddings)

model = SentenceTransformer("google/embeddinggemma-300m")

embeddings = model.encode(s_sentences)

similarities = model.similarity(embeddings, embeddings)

print("distilled similarities")
print(s_similarities)
print("google similarities")
print(similarities)
print("final model similarities")
print(f_similarities)
```
